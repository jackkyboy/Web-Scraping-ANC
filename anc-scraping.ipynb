```python
import nest_asyncio
import asyncio
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeoutError

nest_asyncio.apply()

async def fetch_insurance_data():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)  # Set headless=False to see the browser action
        page = await browser.new_page()
        await page.goto("https://www.ancbroker.com/ประกันภัยรถยนต์")

        try:
            await page.wait_for_selector('div.panel.panel-default', timeout=100000)  # Increased timeout to 60 seconds
        except PlaywrightTimeoutError:
            print("Timeout waiting for selector")
            await browser.close()
            return []

        # Extract the insurance items
        insurance_items = await page.query_selector_all('div.panel.panel-default')

        results = []
        for item in insurance_items:
            company = await item.query_selector('span#ContentPlaceHolder_PackageGroup_CPNAME_0_CPNAME')
            package = await item.query_selector('div[data-title]')
            price = await item.query_selector('span[data-fix]')
            coverage = await item.query_selector('div.panel-body')

            if company and package and price and coverage:
                company_text = await company.inner_text()
                package_text = await package.get_attribute('data-title')
                price_text = await price.inner_text()
                coverage_text = await coverage.inner_text()

                results.append({
                    'company': company_text.strip(),
                    'package': package_text.strip(),
                    'price': price_text.strip(),
                    'coverage': coverage_text.strip()
                })

        await browser.close()
        return results

# Run the fetch_insurance_data function in the event loop
insurance_data = asyncio.run(fetch_insurance_data())

# Print the results
for data in insurance_data:
    print(data)
```

```python
from bs4 import BeautifulSoup

def extract_package_details(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    packages = soup.find_all("div", class_="panel panel-default")
    package_details = []

    for package in packages:
        package_info = {}
        package_name_element = package.find("span", class_="CPNAME")
        if package_name_element:
            package_info["Package Name"] = package_name_element.text.strip()
        else:
            package_info["Package Name"] = "N/A"
        
        premium_info = package.find("span", class_="PREMIUM")
        offer_price_info = package.find("span", class_="OFFER_PRICE")
        
        if premium_info and offer_price_info:
            package_info["Original Price"] = premium_info.text.strip().replace('฿', '').replace('<del>', '').replace('</del>', '')
            package_info["Offer Price"] = offer_price_info.text.strip().replace('฿', '')
        else:
            package_info["Original Price"] = "N/A"
            package_info["Offer Price"] = "N/A"
        
        package_details.append(package_info)

    return package_details

# ตัวอย่างการทดสอบฟังก์ชันด้วย HTML ที่ได้รับ
html_content = """
<div class="panel panel-default" id="GroupCP_VBI">
    <div class="panel-heading panel-heading-link">
        <a class="collapsed" data-toggle="collapse" data-parent="#accordion" href="#VBI">
            <img id="ContentPlaceHolder_PackageGroup_INS_4" class="img-circle" src="../images/ins/VBI.png">
            <span id="ContentPlaceHolder_PackageGroup_CPNAME_4" class="CPNAME">วิริยะประกันภัย</span>
        </a>
    </div>
    <div id="VBI" class="panel-collapse collapse">
        <div class="panel-body package-detail pt-0 pb-0">
            <div class="row line-bt selectProduct" data-id="19736">
                <div class="col-xs-8 col-sm-4 price">
                    <a id="ContentPlaceHolder_PackageGroup_rptPackageList_13_HyperLink1_1" class="button button-default button-pill button-small">
                        <span id="ContentPlaceHolder_PackageGroup_rptPackageList_13_Label3_1" class="PREMIUM">฿<del>22,600</del></span>
                        <span id="ContentPlaceHolder_PackageGroup_rptPackageList_13_Span1_1" class="OFFER_PRICE">฿<span>22,600</span></span>
                    </a>
                </div>
            </div>
        </div>
    </div>
</div>
"""

package_details = extract_package_details(html_content)
for detail in package_details:
    print(detail)
```
